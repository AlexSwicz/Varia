Нам нужно простое выражение, которое удалит названия журналов из бывших колонтитулов.  
Плюс сразу удалим всю пунктуацию, кроме точек, вопросительных и восклицательных знаков и запятых.
И еще латиницу с числительными.

```python
text = txt
text = text.lower()
text = re.sub(r'(вестник спбгу. сер. 9. филология. востоковедение. журналистика. 2016. вып. 4)', '', text)
text = re.sub(r'[a-zA-z]', '', text)
text = re.sub(r'[1234567890]', '', text)
text = re.sub(r'[\(/\)\\\/ç–ãóñöóü;@ä:→«»]', '', text)
#удалим последовательности из точек и запятых, оставшихся от референсов.
text = re.sub(r'( , ,)', '', text)
text = re.sub(r'(  +)', ' ', text)
text = re.sub(r'( \.)', '.', text)
```

В конце статьи может остаться область, состоящая из точек и запятых - обрывки референсов. Удалим их жадным образом.

```python
text = re.sub(r'(\.\.\.).+((\n).+)+', '', text)
```

Будем использовать второй вариант. Создадим функцию, которая будет очищать сразу все файлы.

```python
def preprocessing(text):
    with open(text, encoding = 'utf-8') as fi:
        lo = fi.read() 
        anothtext = lo.lower()
        anothtext = re.sub(r'(вестник спбгу. сер. 9. филология. востоковедение. журналистика. 2016. вып. 4)', '', anothtext)
        anothtext = re.sub(r'(cоциальная справедливость в здравоохранении: опыт и оценки россиян, стр.)', '', anothtext)
        anothtext = re.sub(r'[a-zA-z]', '', anothtext)
        anothtext = re.sub(r'[1234567890]', '', anothtext)
        anothtext = re.sub(r'[\(ʹ/\)\\\/ńç–ãóñöʃóü;í@ä:→«»àšīžāʲɨøŋɔɣəʔśəəʔìďʔčīēāîéáúáèééáœƒﬂõąłłżęъőхąłłę]', '', anothtext)
        anothtext = re.sub(r'(references).+(.+)+', '', anothtext)
        anothtext = re.sub(r'удк', '', anothtext)
        words = re.findall('\w+', anothtext)
        newtext = ' '.join(words)
        newtext = re.sub(r'(вестник санкт петербургского университета сер вып языкознание)', '', newtext)
        newtext = re.sub(r'(вестник московского университета серия психология)', '', newtext)
        newtext = re.sub(r'(институт славяноведения ран)', '', newtext)    
        newtext = re.sub(r'(московский педагогический государственный университет)', '', newtext)   
        newtext = re.sub(r'(вестник томского государственного университета)', '', newtext)   
        newtext = re.sub(r'(российская федерация москва)', '', newtext)
        newtext = re.sub(r'(вестник спбгу сер вып)', '', newtext)
        newtext = re.sub(r'(томский журнал линг и антр)', '', newtext)   
        newtext = re.sub(r'(работа над статьей велась в рамках проекта рнф)', '', newtext)  
    return newtext
```

```python
if __name__ == '__main__':
    os.mkdir('LingPrep')
    for d,file in enumerate(os.listdir('C:\\Users\\asus\\Desktop\\CleanedLinguistics')):
        with open("LingPrep\\{0}.txt".format(d),'w',encoding='utf-8') as f:
            texttext = preprocessing('C:\\Users\\asus\\Desktop\\CleanedLinguistics\\{0}'.format(file))
            f.write(texttext)
```
